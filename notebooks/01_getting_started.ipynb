## Pyravelry

---

Pyravelry is a Python library for interacting with the Ravelry API. This notebook will show the basics for how to interact with the library.

```python
%load_ext autoreload
%autoreload 2
```

```python
from pyravelry import Client, Settings
```

```python
# need to define a .env with RAVELRY_USERNAME and RAVELRY_API_KEY
# or define it on the command line.
settings = Settings()
client = Client(settings=settings)
```

## Getting data

Data from an API call is cached by default (using [hishel](https://hishel.com/1.0/)). The data is returned as a pydantic model, but it can be converted to `pandas` or `polars` dataframes using properties of the custom `ravelry.models.base.BaseRavelryModel` pydantic model. 

```python
results = client.search.query(query="merino", limit=50, types=["Yarn"])
```

```python
results.pandas.head()
```

```python
results.polars.head()
```
